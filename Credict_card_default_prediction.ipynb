```python
df.rename(columns = {"default.payment.next.month":"default"}, inplace = True)
```

```python
# df["default"].unique()
# default.payment.next.month = Will the customer miss (default on) next month’s payment?
df["default"].value_counts(normalize = True)
```

```python
bill_amt = [f"BILL_AMT{i}" for i in range(1,7)]
df["avg_bill_amt"] = df[bill_amt].mean(axis = 1)
```

```python
pay_amt = [f"PAY_AMT{i}" for i in range(1,7)]
df["avg_pay_amt"] = df[pay_amt].mean(axis=1)
```

```python
delay_amt = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
df["max_delay"] = df[delay_amt].max(axis=1)
```

```python
df["payment_ratio"] = df["avg_pay_amt"]/(df["avg_bill_amt"]+1)
```

```python
df.head()
```

**Model training**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
# train and test split
x = df.drop("default" ,axis =1)
y = df["default"]
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size =0.2,stratify = y, random_state = 42)
```

```python
# Column transform
num_features = x.columns

col_trans = ColumnTransformer(
    transformers=[
        ("scaler",StandardScaler(),num_features)
    ]
)
```

```python
# Baseline Model (random forest)
rf_pipe = Pipeline([
    ("col_transform",col_trans),
    ("model_rf",RandomForestClassifier(
        n_estimators=300,
        max_depth = 10,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    ))
])

rf_pipe.fit(x_train,y_train)
```

```python
# Advanced Model XGBoost
scal_pos_weight = df["default"].value_counts()[0] / df["default"].value_counts()[1]

xgb_pipe = Pipeline([
    ("col_transform",col_trans),
    ("xgb",XGBClassifier(
        n_estimators = 300,
        max_depth = 10,
        scale_pos_weight = scal_pos_weight,
        learning_rate = 0.05,
        subsample = 0.8,
        colsample_bytree = 0.8,
        random_state = 42,
        eval_metrics = "logloss"
    ))

])

xgb_pipe.fit(x_train,y_train)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import RandomizedSearchCV

cv = StratifiedKFold(n_splits = 5, shuffle = True, random_state=42)

param_grid = {
    "xgb__n_estimators":[300,500,800],
    "xgb__learning_rate": [0.01, 0.05, 0.1],
    "xgb__max_depth": [3, 5, 7],
    "xgb__subsample": [0.7, 0.8, 1.0],
    "xgb__colsample_bytree": [0.7, 0.8, 1.0]
}

search = RandomizedSearchCV(
    xgb_pipe,
    param_distributions = param_grid,
    n_iter = 30,
    scoring = "roc_auc",
    cv= cv,
    n_jobs = 1,
    verbose = 2,
    random_state = 42
)

search.fit(x_train,y_train)

best_model = search.best_estimator_
```

**Model Evaluation**

```python
from sklearn.metrics import classification_report, roc_auc_score
```

```python
def evaluate_train(model):
  preds = model.predict(x_train)
  probs = model.predict_proba(x_train)[:,1]

  print("Classification Report \n",classification_report(y_train,preds))
  print("ROC-AUC Score \n",roc_auc_score(y_train,probs))

print("random Forest")
evaluate_train(rf_pipe)
print("\n")
print("\n")

print("XGBoost")
evaluate_train(xgb_pipe)
print("\n")
print("\n")

print("Best Model")
evaluate_train(best_model)
```

```python
def evaluate_test(model):
  preds = model.predict(x_test)
  probs = model.predict_proba(x_test)[:,1]

  print("Classification report", classification_report(y_test,preds))
  print("ROC AUC Score",roc_auc_score(y_test,probs))

print("Random Forest")
evaluate_test(rf_pipe)
print("\n")
print("\n")

print("XGBoost")
evaluate_test(xgb_pipe)
print("\n")
print("\n")

print("Best Model")
evaluate_train(best_model)
```

```python
# Threshold Optimization
from sklearn.metrics import f1_score

probs = best_model.predict_proba(x_test)[:,1]

threshold = np.arange(0.2,0.8, 0.05)
scores = [f1_score(y_test, probs>t) for t in threshold]

best_threshold = threshold[np.argmax(scores)]
best_threshold
```
